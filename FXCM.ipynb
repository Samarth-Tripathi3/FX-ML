```python
import datetime as dt
import numpy as np
import pandas as pd
import fxcm
api = fxcm.fxcm(config_file='fxcm.cfg')
```

```python
data = pd.DataFrame(candles[['askclose', 'bidclose']].mean(axis=1),
                    columns=['midclose'])
```

```python
data.head()
```

```python
data.tail()
```

```python
import cufflinks as cf  # Cufflinks
cf.set_config_file(offline=True)  # set the plotting mode to offline
data.iplot()
```

```python
data['returns'] = np.log(data / data.shift(1))
```

```python
lags = 3
cols = []
for lag in range(1, lags + 1):
    col = 'lag_%s' % lag
    data[col] = data['returns'].shift(lag)
    cols.append(col)
```

```python
col = 'momentum'
data[col] = data['returns'].rolling(5).mean().shift(1)
cols.append(col)
```

```python
cols
```

```python
from pylab import plt
plt.style.use('seaborn')
%matplotlib inline
```

```python
data['direction'] = np.sign(data['returns'])
to_plot = ['midclose', 'returns', 'direction']
#data[to_plot].iloc[:100].plot(figsize=(10, 6),subplots=True, style=['-', '-', 'ro'], title='EUR/USD');
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
nos = 50
mu = data['returns'].rolling(nos).mean()
v = data['returns'].rolling(nos).std()
def digits(s):
    r = np.where(s < mu - v, 0, np.nan)
    r = np.where(s > mu - v, 1, r)
    r = np.where(s > mu, 2, r)
    r = np.where(s > mu + v, 3, r)
    return r  
```

```python
data[cols].apply(digits)
bins = [mu - v, mu, mu + v]
train_x, test_x, train_y, test_y = train_test_split(
    data[cols].apply(digits).dropna(),                   
    np.sign(data['returns']).iloc[nos-1:],
    test_size=0.50, random_state=111)


####### Changed from 0.5 -> 0.4
#data[cols].apply(digits)
#bins = [mu - v, mu, mu + v]
#train_x, test_x, train_y, test_y = train_test_split(
#    data[cols].apply(digits).dropna(),                   
#    np.sign(data['returns']).iloc[nos-1:],
#    test_size=0.50, random_state=111)
```
